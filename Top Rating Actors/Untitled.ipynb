```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
top_actors_rating = pd.read_csv('top_actors_rating.csv')
```

```python
genre_stats = (
    top_actors_rating.groupby(['actor_name', 'genre'])
    .agg(avg_rating=('movie_rating', 'mean'), count=('movie_rating', 'count'))
    .reset_index()
)
```

```python
genre_stats
```

```python
max_count = genre_stats.groupby('actor_name')['count'].transform('max')
max_count
```

```python
top_genres = genre_stats[genre_stats['count'] == max_count]
top_genres
```

```python
max_avg_rating = top_genres.groupby('actor_name')['avg_rating'].transform('max')
max_avg_rating
```

```python
top_genres = top_genres[top_genres['avg_rating'] == max_avg_rating]
top_genres
```

```python
top_genres = top_genres.sort_values(['actor_name', 'genre']).drop_duplicates(
    'actor_name'
)
```

```python
top_genres['rank'] = top_genres['avg_rating'].rank(method='dense', ascending=False)
```

```python
top_genres
```

```python
max_rank = top_genres[top_genres['rank'] <= 3]['rank'].max()
max_rank
```

```python
result = (
    top_genres[top_genres['rank'] <= max_rank]
    .loc[:, ['actor_name', 'genre', 'avg_rating']]
    .sort_values('avg_rating', ascending=False)
    .reset_index(drop=True)
)
```
