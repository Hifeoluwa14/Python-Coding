```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
# load csv file into dataframe
top_actors_rating = pd.read_csv('top_actors_rating.csv')

# show dataframe
top_actors_rating
```

Find the top actors based on their average movie rating within the genre they appear in most frequently.
•  For each actor, determine their most frequent genre (i.e., the one they’ve appeared in the most).
•   If there is a tie in genre count, select the genre where the actor has the highest average rating.
•   If there is still a tie in both count and rating, include all tied genres for that actor.


Rank the resulting actor + genre pairs by their average rating in descending order.
•  Return all actor + genre pairs that fall within the top 3 ranks (not rows), including ties.
•  Do not skip ranks — for example, if multiple actors are tied at rank 1, the next rank is 2.

```python
# sort by actors name and genre
# added a column that shows the average rating of each genre
# added another column that count the apperance of actor for each genre
top_actors_rating =  top_actors_rating[['actor_name','genre','movie_rating']].sort_values(['actor_name','genre']).groupby(['actor_name','genre'], as_index = False).agg(avg_rating=('movie_rating', 'mean'), apperance=('movie_rating', 'count'))

# show dataframe
top_actors_rating
```

Find the top actors based on their average movie rating within the genre they appear in most frequently. • For each actor, determine their most frequent genre (i.e., the one they’ve appeared in the most). • If there is a tie in genre count, select the genre where the actor has the highest average rating. • If there is still a tie in both count and rating, include all tied genres for that actor.

```python
# add column that show maximum apperance by each actor
top_actors_rating['max_apperance'] = top_actors_rating.groupby(['actor_name'])['apperance'].transform('max')

# show table
top_actors_rating
```

```python
# Find the top actors based on their average movie rating within the genre they appear in most frequently

# This is done by returing rows where apperance and max_apperance matches
top_actors_rating = top_actors_rating[['actor_name', 'avg_rating', 'genre', 'apperance']][top_actors_rating['apperance' ] == top_actors_rating['max_apperance']]

#show
top_actors_rating
```

```python
#If there is a tie in genre count, select the genre where the actor has the highest average rating
#If there is still a tie in both count and rating, include all tied genres for that actor.

# Added a column that shows max avg_rating of each actor
top_actors_rating['max_avg_rating'] = top_actors_rating.groupby(['actor_name'])['avg_rating'].transform('max')

#show
top_actors_rating 
```

```python
#If there is a tie in genre count, select the genre where the actor has the highest average rating
#If there is still a tie in both count and rating, include all tied genres for that actor.

# This is done by returing rows where avg_rating and max_avg_rating matches
top_actors_rating = top_actors_rating[['actor_name', 'avg_rating', 'genre', 'apperance']][top_actors_rating['avg_rating' ] == top_actors_rating['max_avg_rating']]

# show
top_actors_rating
```

Rank the resulting actor + genre pairs by their average rating in descending order. • Return all actor + genre pairs that fall within the top 3 ranks (not rows), including ties. • Do not skip ranks — for example, if multiple actors are tied at rank 1, the next rank is 2.

```python
#Rank the resulting actor + genre pairs by their average rating in descending order.
top_actors_rating['rank'] = top_actors_rating['avg_rating'].rank(method = 'dense', ascending = False)

top_actors_rating
```

```python
#Return all actor + genre pairs that fall within the top 3 ranks (not rows), including ties.
top_actors_rating[['actor_name', 'genre','avg_rating']][top_actors_rating['rank'] <= 3]
```
