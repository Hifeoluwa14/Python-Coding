```python
import pandas as pd
```

```python
# load csv file
customers = pd.read_csv('customers.csv')

#show
customers
```

```python
#load csv into dataframe
orders = pd.read_csv('orders.csv')

#show
orders
```

Find the percentage of shipable orders.
Consider an order is shipable if the customer's address is known.

```python
# add all orders and sum all cost by user customer id
customer_order = orders[['cust_id', 'order_details', 'total_order_cost']].groupby(['cust_id'], as_index = False).agg(no_of_order =('order_details','count'), total_cost =('total_order_cost','sum'))

#show
customer_order
```

```python
# match customer's address to customer_order 
customer_order = pd.merge(customer_order,customers[['address', 'id']], left_on = 'cust_id' , right_on ='id')

# show
customer_order
```

```python
# fill NaN in address column with 'unknown address'
customer_order['address'] = customer_order['address'].fillna('unknown address')

#show
customer_order
```

```python
# divide no of order of unknown address by total no of order then multiply by 100 and round up to 2 decimal place
shippable_order =round(customer_order[['no_of_order']][customer_order['address'] != 'unknown address'].sum() / customer_order[['no_of_order']].sum() * 100,2)

# show result
shippable_order 
```

```python
# add '%' to result
shippable_order_percentage = f'{shippable_order}%'

#show result
shippable_order_percentage
```
