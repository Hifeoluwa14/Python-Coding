```python
import numpy as np
import pandas as pd
```

```python
# load customers csv file to dataframe
customers = pd.read_csv('customers.csv')

#show
customers
```

```python
# load orders csv into dataframe
orders = pd.read_csv('orders.csv')

#show
orders
```

Find the customer with the highest daily total order cost between 2019-02-01 to 2019-05-01. If customer had more than one order on a certain day, sum the order costs on daily basis.Output customer's first name, total cost of their items, and the date.


For simplicity, you can assume that every first name in the dataset is unique.

```python
#merge customers and order into single dataframe
customer_order = pd.merge(customers,orders,left_on='id',right_on = 'cust_id')

#check data consistency and datatype for columns
customer_order.info()
```

```python
# change order_date datatype to datetime
customer_order['order_date'] = pd.to_datetime(customer_order['order_date'])

#verify
customer_order.info()
```

```python
# filter date to needed timeline and also sum up daily transactions by each user
customer_order = customer_order[['cust_id','first_name','last_name','order_date','total_order_cost']][customer_order['order_date'].between('2019-02-01', ' 2019-05-01' )].groupby(['cust_id','first_name','last_name','order_date'], as_index=  False).sum()

#show
customer_order
```

```python
# Return row where total order cost matches the maximum total order cost
customer_order[['first_name','order_date','total_order_cost']][customer_order['total_order_cost'] == customer_order['total_order_cost'].max()]
```
