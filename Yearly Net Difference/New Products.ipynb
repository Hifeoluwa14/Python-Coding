```python
import pandas as pd
import numpy as np
```

```python
#load csv into dataframe
car_launches = pd.read_csv('new_products.csv')

#show
car_launches
```

Calculate the net change in the number of products launched by companies in 2020 compared to 2019. Your output should include the company names and the net difference.
(Net difference = Number of products launched in 2020 - The number launched in 2019.)

```python
#create a dataframe from car launches that return company_name, year and number of products for each year
no_of_cars = car_launches[['company_name','year','product_name']].groupby(['company_name','year'], as_index = False).agg(product_no=('product_name', 'count'))

#show
no_of_cars
```

```python
# create dataframe that shows all company with only 2019 car production
car_launches_2019 = no_of_cars[['company_name','year','product_no']][no_of_cars['year'] == 2019]

#show
car_launches_2019
```

```python
# create dataframe that shows all company with only 2019 car production
car_launches_2020 = no_of_cars[['company_name','year','product_no']][no_of_cars['year'] == 2020]

#show
car_launches_2020
```

```python
#merge both dataframe together
car_launches_by_year = pd.merge(car_launches_2019,car_launches_2020, left_on = 'company_name', right_on = 'company_name')

#show
car_launches_by_year
```

```python
#create a column that get the net difference of car produced in 2020 and 2019
car_launches_by_year['net_difference'] = car_launches_by_year['product_no_y'] - car_launches_by_year['product_no_x']

#show
car_launches_by_year
```

```python
#show company_name and net_difference in the number of products launched by companies in 2020 compared to 2019
car_launches_by_year[['company_name','net_difference']]
```
