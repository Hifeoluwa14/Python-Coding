```python
import pandas as pd
```

```python
# load csv into dataframe
fb_active_users = pd.read_csv('fb_active_users.csv')

#show
fb_active_users
```

```python
# load csv into dataframe
fb_comments_count = pd.read_csv('fb_comments_count.csv')

#show
fb_comments_count
```

Which countries have risen in the rankings based on the number of comments between Dec 2019 vs Jan 2020? Hint: Avoid gaps between ranks when ranking countries

```python
# check dataframe details
fb_comments_count.info()
```

```python
# change created_at datatype to datetime
fb_comments_count['created_at'] = pd.to_datetime(fb_comments_count['created_at'])

#show
fb_comments_count.info()
```

```python
# extract only comments made in december
dec_comments = fb_comments_count[['user_id', 'created_at', 'number_of_comments']][fb_comments_count['created_at'].between('2019-12-01', '2019-12-31')]

#show
dec_comments 
```

```python
# sum comments for each user
dec_comments = dec_comments[['user_id', 'number_of_comments']].groupby(['user_id'], as_index= False).sum()

#show
dec_comments
```

```python
# extract only comments made in january
jan_comments = fb_comments_count[['user_id', 'created_at', 'number_of_comments']][fb_comments_count['created_at'].between('2020-01-01', '2020-01-31')]

#show
jan_comments 
```

```python
# sum comments for each user
jan_comments = jan_comments[['user_id', 'number_of_comments']].groupby(['user_id'], as_index= False).sum()

#show
jan_comments
```

```python
# extract country of users
users_country = fb_active_users[['user_id','country']].sort_values('user_id')

#show
users_country
```

```python
# match country of each user
dec_users = pd.merge(dec_comments, users_country, left_on = 'user_id', right_on = 'user_id')

#show
dec_users
```

```python
# match country of each user
jan_users = pd.merge(jan_comments, users_country, left_on = 'user_id', right_on = 'user_id')

#show
jan_users
```

```python
# sum comments by country of users
dec_comment_by_country = dec_users[['country','number_of_comments']].groupby(['country'], as_index = False).sum()

#show
dec_comment_by_country
```

```python
# sum comments by country of users
jan_comment_by_country = jan_users[['country','number_of_comments']].groupby(['country'], as_index = False).sum()

#show
jan_comment_by_country
```

```python
# add a column that ranks countries by number of comments 
dec_comment_by_country['rank'] = dec_comment_by_country['number_of_comments'].rank(method = 'dense', ascending = False)

#show
dec_comment_by_country
```

```python
# add a column that ranks countries by number of comments 
jan_comment_by_country['rank'] = jan_comment_by_country['number_of_comments'].rank(method = 'dense', ascending = False)

#show
jan_comment_by_country
```

```python
# match both months table by country
country_rank = pd.merge(dec_comment_by_country, jan_comment_by_country, left_on = 'country', right_on = 'country')

#show
country_rank
```

```python
# rename each months rank column
country_rank = country_rank.rename(columns = {'rank_x':'dec_rank', 'rank_y': 'jan_rank'})

#show
country_rank
```

```python
#check and return countries that has ranked higher in Jan compare to their dec rank
country_rank[['country']][country_rank['jan_rank'] < country_rank['dec_rank']]
```
