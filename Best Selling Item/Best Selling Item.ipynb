```python
import pandas as pd
import numpy as np
```

```python
# import csv file into dataframe
online_retail = pd.read_csv('online_retail.csv')
```

Find the best-selling item for each month (no need to separate months by year). The best-selling item is determined by the highest total sales amount, calculated as: total_paid = unitprice * quantity. Output the month, description of the item, and the total amount paid.

```python
# show
online_retail
```

```python
# create a column that returns the total sales
online_retail['total_sale'] = online_retail['quantity'] * online_retail['unitprice']

#show
online_retail
```

```python
#split the invoice date column
date = online_retail['invoicedate'].str.split(pat = '/', expand= True)

#show
date
```

```python
#merge the month column to the online_retail dataframe
online_retail = pd.merge(online_retail, date[1], left_index = True, right_index = True)

#show
online_retail
```

```python
# rename column
online_retail = online_retail.rename(columns = {1:'month'})

#show
online_retail
```

```python
# return monthly sales by summing each item sales for each month
monthly_sales = online_retail[['month','stockcode','description','total_sale']].groupby(['month','stockcode','description' ], as_index = False).sum()

#show
monthly_sales
```

```python
# create a column that returns max sales of each month and attach it to the specific month
monthly_sales['max'] = monthly_sales.groupby(['month'])['total_sale'].transform('max')

#show
monthly_sales
```

```python
# return rows where the max column matches the total sales 
best_selling_item = monthly_sales[['month','description', 'total_sale']][monthly_sales['total_sale'] == monthly_sales['max']]

#show
best_selling_item
```
