```python
import pandas as pd 
#https://platform.stratascratch.com/coding/9911-departments-with-5-employees?code_type=2
```

```python
#load csv file into dataframe
employee = pd.read_csv('employee.csv')

#show
employee
```

Find departments with more than or equal 5 employees.

```python
#count number of employees for each department
employee = employee[['id','employee_title', 'department' ]].groupby(['department', 'employee_title'], as_index = False).agg(no_of_employee = ('id','count'))

#show
employee
```

```python
#find department with more than or equals 5 employees
employee['department'][employee['no_of_employee'] >= 5]
```
