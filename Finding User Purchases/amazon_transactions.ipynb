Identify returning active users by finding users who made a second purchase within 1 to 7 days after their first purchase. Ignore same-day purchases. Output a list of these user_ids.

```python
#Check dataframe details for null values and datatype
amazon_transactions.info()
```

```python
#change created_at datatype to datetime
amazon_transactions['created_at'] = pd.to_datetime(amazon_transactions['created_at'])

#show
amazon_transactions.info()
```

```python
#sort table needed to perform task
amazon = amazon_transactions[['user_id', 'created_at']].sort_values(['user_id','created_at'])

#show
amazon
```

```python
#drop duplicate rows
amazon.drop_duplicates(inplace = True)
```

```python
#iterate through dataframe to check if a user makes another purchase with 7 days
user_id = []
current_id = None
first_date = None

for index, row in amazon.iterrows():
    if row['user_id'] != current_id:

        current_id = row['user_id']
        first_date = row['created_at']
        continue
    if (row['created_at'] - first_date).days <= 7:
        user_id.append(row['user_id'])
        
#Tranform user_id list to set to return just unique ids, then transform set back to list
user = list(set(user_id))

#shows the list
user
```
